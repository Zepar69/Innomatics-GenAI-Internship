```python
#Mobile recharge validation system

def validate_recharge(amount):
    valid_plans=[199,299,399,599]
    if amount<50:
        return False,"Amount must be 50 or more."
    if amount not in valid_plans:
        return False,"Invalid plan. Choose from 199,299,399,599."
    return True,"Recharge successful!"
while True:
    amount=int(input("Enter recharge amount: "))
    status,message=validate_recharge(amount)
    print(message)
    if status:
        break
```

```python
#Inventory reorder alert system

def check_inventory(inventory):
    for product,stock in inventory.items():
        if stock<15:
            print(f"{product}:Reorder Alert")
        else:
            print(f"{product}:Stock OK")
inventory={"Rice":10,"Sugar":20,"Oil":5,"Salt":30}
check_inventory(inventory)
```

```python
#Student result processing

def process_result(marks):
    total=0
    for m in marks:
        total+=m
    average=total/len(marks)
    print(f"Average Marks:{average:.2f}")
    if average>=50:
        return "Pass"
    else:
        return "Fail"
marks=[]
n=int(input("Enter number of subjects: "))
for i in range(n):
    mark=float(input(f"Enter marks for subject {i+1}: "))
    marks.append(mark)
result=process_result(marks)
print("Result:",result)
```

```python
#Cab fare estimator with retry option

def calculate_fare(distance,peak_hour):
    base_fare=50
    fare=base_fare+(distance*12)
    if peak_hour:
        fare+=fare*0.25
    return fare
while True:
    distance=float(input("Enter distance in km:"))
    peak=input("Is it peak hour? (yes/no):").strip().lower()
    peak_hour=peak=="yes"
    fare=calculate_fare(distance,peak_hour)
    print(f"Estimated Fare:{fare:.2f}")
    retry=input("Calculate again? (yes/no):").strip().lower()
    if retry!="yes":
        break
```

```python
#Employee attendace eligibility checker

def check_eligibility(attendance):
    present=0
    for day in attendance:
        if day=="P":
            present+=1
    percentage=(present/len(attendance))*100
    print(f"Attendance:{percentage:.2f}%")
    if percentage>=75:
        return "Eligible"
    else:
        return "Not Eligible"
attendance=[]
n=int(input("Enter number of days: "))
for i in range(n):
    status=input(f"Day{i+1}-Present or Absent? (P/A): ").strip().upper()
    attendance.append(status)
result=check_eligibility(attendance)
print("Status:",result)
```

```python
#Password strength checker

def check_password(password):
    special_chars="@#$"
    has_digit=False
    has_special=False
    if len(password)<8:
        return "Your password is too short!"
    for char in password:
        if char.isdigit():
            has_digit=True
        if char in special_chars:
            has_special=True
    if not has_digit:
        return "Weak: Password must contain at least one digit."
    if not has_special:
        return "Weak: Password must contain @, #, or $."
    return "Strong Password!"
password=input("Enter password:")
print(check_password(password))
```
