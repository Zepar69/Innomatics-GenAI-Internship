```python
#Prob 1-Smart parking lot management system

def parking_lot():
    capacity=int(input("Enter Parking Capacity:"))
    n=int(input("How many log entries?"))
    logs=[]
    for i in range(n):
        log=input(f"Entry {i+1}-(IN/OUT):").strip().upper()
        logs.append(log)
    count=0
    for log in logs:
        if log=="IN":
            count+=1
        elif log=="OUT":
            count-=1
    print(f"\nCurrently Parked Vehicles:{count}")
    if count>=capacity:
        print("Parking Status:FULL-Capacity Exceeded!")
    else:
        print("Parking Status:Available")
parking_lot()
```

```python
#Prob 2-Online food delivery time estimator

def delivery_time():
    distance=float(input("Enter Distance(km):"))
    traffic=input("Enter Traffic Level (Low/Medium/High):").strip().capitalize()
    weather=input("Enter Weather (Clear/Foggy/Rainy):").strip().capitalize()
    base_time=distance*3
    if traffic=="High":
        base_time+=20
    elif traffic=="Medium":
        base_time+=10
    if weather=="Rainy":
        base_time+=15
    elif weather=="Foggy":
        base_time+=10
    print(f"\nEstimated Delivery Time:{int(base_time)} minutes")
delivery_time()
```

```python
#Prob 3-Movie theatre seat occupancy analyzer

def theatre_occupancy():
    total_seats=int(input("Enter Total Seats:"))
    booked=int(input("Enter Number of Booked Seats:"))
    occupancy=(booked/total_seats)*100
    print(f"\nOccupancy:{occupancy:.0f}%")
    if occupancy==100:
        print("Show Status:Housefull")
    elif occupancy>=75:
        print("Show Status:Almost Full")
    else:
        print("Show Status:Available")
theatre_occupancy()
```

```python
#Prob 4-Cloud server Classification System

def server_load(cpu_readings):
    avg=sum(cpu_readings)/len(cpu_readings)
    print(f"Average CPU Load:{avg:.0f}%")
    if avg<50:
        print("Server Status:Normal")
    elif avg<=80:
        print("Server Status:Warning")
    else:
        print("Server Status:Critical")
server_load([40,69,67,95,75])
```

```python
#Prob 5-Smart classroom resourse usage monitor

def resource_monitor():
    threshold=8
    n=int(input("Enter number of resources:"))
    usage={}
    for i in range(n):
        resource=input(f"Enter Resource {i+1} name:").strip()
        hours=float(input(f"Enter usage hours for {resource}:"))
        usage[resource]=hours
    overused=[]
    for resource,hours in usage.items():
        if hours>threshold:
            overused.append(resource)
    if overused:
        print(f"\nOverused Resources:{', '.join(overused)}")
        print("Energy Alert:Yes")
    else:
        print("\nAll resources are within limits")
        print("Energy Alert:No")
resource_monitor()
```

```python
#Online event registration capacity controller

def event_registration():
    capacity=int(input("Enter Event Capacity:"))
    registrations=int(input("Enter Number of Registrations:"))
    confirmed=min(registrations,capacity)
    waitlisted=max(0,registrations-capacity)
    status="Closed" if registrations>=capacity else "Open"
    print(f"\nConfirmed Registrations:{confirmed}")
    print(f"Waitlisted Users:{waitlisted}")
    print(f"Registration Status:{status}")
event_registration()
```
