```python
#User login check

user="admin"
pwd="1234"

user_inp=input("Enter username:")
user_pass=input("Enter password:")

if user_inp == user and user_pass == pwd:
    print("Successful login!")
else:
    print("Invalid username/password!")
```

```python
#Pass/ Fail check

marks=[67,69,33,27,21]
failed=0
passed=0

for mark in marks:
    if mark>=50:
        passed=passed+1
    else:
        failed=failed+1
        
print("Number of students passed:",passed)
print("Number of students failed:",failed)
```

```python
#Simple Data Cleaner

names=["ABC", "EFG ", " HIGJLM "]

cleaned_names = []

for n in names:
    cleaned=n.strip().lower()
    cleaned_names.append(cleaned)
    
print(cleaned_names)
```

```python
#Message Length Analyzer

messages=["Hello","We welcome you to the platform", "Ok"]

for msg in messages:
    length=len(msg)
    print(f"Message:'{msg}'has length:{length}")
    
    if length>10:
        print("This message is over 10 characters!")
```

```python
#Error Message Detector

logs= ["INFO", "ERROR", "WARNING", "ERROR"]
err_count=0

for log in logs:
    if log=="ERROR":
        err_count=err_count+1
        
print("Total number of errors:", err_count) 
```
