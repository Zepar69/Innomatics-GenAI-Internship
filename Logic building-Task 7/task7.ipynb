```python
#Smart light controller

class SmartLight:
    def __init__(self,name,action):
        self.name=name
        self.action=action
    def display(self):
        print(f"{self.name} is {self.action}")
name=input("Light Name:")
action=input("Action:")
light=SmartLight(name,action)
light.display()
```

```python
#Employee ID card system

class Employee:
    def __init__(self,name,emp_id,department):
        self.name=name
        self.emp_id=emp_id
        self.department=department
    def display(self):
        print("Employee ID Card")
        print(f"Name:{self.name}")
        print(f"ID:{self.emp_id}")
        print(f"Department:{self.department}")
name=input("Employee Name:")
emp_id=input("Employee ID:")
department=input("Department:")
emp=Employee(name,emp_id,department)
emp.display()
```

```python
#Mobile contact record

class Contact:
    def __init__(self,name,phone):
        self.name=name
        self.phone=phone
    def display(self):
        print("Contact Saved")
        print(f"Name:{self.name}")
        print(f"Phone:{self.phone}")
name=input("Contact Name:")
phone=input("Phone Number:")
contact=Contact(name,phone)
contact.display()
```

```python
#Product price tag generator

class Product:
    def __init__(self,name,price):
        self.name=name
        self.price=price
    def display(self):
        print(f"Product:{self.name}")
        print(f"Price:{self.price}")
name=input("Product Name:")
price=input("Price:")
product=Product(name,price)
product.display()
```

```python
#Movie rating display system

class Movie:
    def __init__(self,name,rating):
        self.name=name
        self.rating=rating
    def display(self):
        print(f"Movie:{self.name}")
        print(f"Rating:{self.rating}/5")
name=input("Movie Name:")
rating=input("Rating:")
movie=Movie(name,rating)
movie.display()
```

```python
#Deliver address manager

class Delivery:
    def __init__(self,customer,address):
        self.customer=customer
        self.address=address
    def display(self):
        print("Delivery Details")
        print(f"Customer:{self.customer}")
        print(f"Address:{self.address}")
customer=input("Customer Name:")
address=input("Address:")
delivery=Delivery(customer,address)
delivery.display()
```
