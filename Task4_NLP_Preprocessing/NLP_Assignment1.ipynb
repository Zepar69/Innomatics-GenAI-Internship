## Task 1 – Conceptual Understanding

**Q1. What is the difference between "Love" and "love" in NLP?**

In NLP, text is case-sensitive by default, which means "Love" and "love" are treated as completely different words. This can mess things up because it increases the number of unique words in the dataset and makes the model less efficient. To avoid this, we usually convert everything to lowercase during preprocessing so both "Love" and "love" are treated the same.

---

**Q2. What happens if stopwords are not removed?**

If we don't remove stopwords, the dataset ends up full of common words like "is", "the", and "and" that don't really add much meaning. This makes the data more bulky and noisy. Due to which, models take more time to process the data and may not perform as well since they're focusing on useless words instead of meaningful ones.

---

**Q3. Two real-world scenarios where removing stopwords can be harmful:**

1. In sentiment analysis, removing words like "not" can completely change the meaning. For example, "not good" would become just "good". This is obviously wrong!

2. In question-answer systems, removing words such as who, what or where takes away the main intent of the question making it harder for a model to understand what's being asked.

---

**Q4. Difference between Stemming and Lemmatization:**

**Stemming:** A rule-based approach where we remove prefixes or suffixes to get to some base form of a word without considering its context.
- Examples: "running" → "run", "studies" → "studi" (may not be a real word)

**Lemmatization:** An advanced process that takes into account the context of a word before converting it into its actual dictionary base form so that the output is always a proper word.
- Examples: "running" → "run", "studies" → "study", "better" → "good"

```python
import re
from collections import Counter

def preprocess_text(text):
    try:
        # Input validation – return empty if not a string or blank
        if not isinstance(text, str) or text.strip() == "":
            return [], ""

        # Lowercase
        text = text.lower()

        # Remove URLs and email-like patterns
        text = re.sub(r"http\S+|www\S+|https\S+|\S+@\S+", "", text)

        # Remove numbers
        text = re.sub(r"\d+", "", text)

        # Handle repeated characters (e.g., soooo → s, baaad → bad)
        text = re.sub(r'(.)\1{2,}', r'\1', text)

        # Remove punctuation and special characters
        text = re.sub(r'[^a-zA-Z\s]', '', text)

        # Remove extra whitespace
        text = re.sub(r'\s+', ' ', text).strip()

        # Tokenize
        tokens = text.split()

        # Remove very short tokens but keep 'no' and 'not'
        tokens = [word for word in tokens if len(word) > 2 or word in ['no', 'not']]

        clean_sentence = " ".join(tokens)

        return tokens, clean_sentence

    except Exception as e:
        return [], ""

print("preprocess_text() defined successfully.")
```

```python
test_sentences = [
    "Get 100% FREE access now!!!",
    "I absolutely looooved this product 😍😍",
    "Worst service ever... 0/10",
    "Call me at 9876543210",
    "This is THE best course!!!",
    "Visit https://openai.com now!",
    "Nooooo this is baaad!!!",
    "OK OK OK I got it",
    "Win $$$ now!!! Limited offer!!!",
    "I am not happy with this"
]

for sentence in test_sentences:
    tokens, clean = preprocess_text(sentence)
    print("\nOriginal:", sentence)
    print("Tokens:", tokens)
    print("Clean Sentence:", clean)
```

```python
def token_analysis(tokens):
    if len(tokens) == 0:
        return 0, 0, 0

    total   = len(tokens)
    unique  = len(set(tokens))
    avg_len = sum(len(word) for word in tokens) / total

    return total, unique, round(avg_len, 2)


analytics_results = []

for sentence in test_sentences:
    tokens, _ = preprocess_text(sentence)
    total, unique, avg = token_analysis(tokens)
    analytics_results.append({
        "sentence": sentence,
        "tokens":   tokens,
        "total":    total
    })
    print("\nSentence:", sentence)
    print("Total:", total, "| Unique:", unique, "| Avg Length:", avg)


# Analysis Questions
most_noisy    = min(analytics_results, key=lambda x: x['total'])
most_retained = max(analytics_results, key=lambda x: x['total'])

print("\n\nAnalysis Questions:")
print("\nQ: Which sentence had the most noise?")
print(f"A: \"{most_noisy['sentence']}\"")
print(f"   Only {most_noisy['total']} token(s) retained — most words were noise.")

print("\nQ: Which sentence retained the most meaningful tokens?")
print(f"A: \"{most_retained['sentence']}\"")
print(f"   Retained {most_retained['total']} tokens — contained the most meaningful words.")
```

```python
from collections import Counter

all_tokens = []

for sentence in test_sentences:
    tokens, _ = preprocess_text(sentence)
    all_tokens.extend(tokens)

counter = Counter(all_tokens)

print("Top 10 most frequent words:")
print(counter.most_common(10))

print("\nTop 5 least frequent words:")
print(counter.most_common()[-5:])
```

```python
def full_pipeline(text_list):
    all_tokens      = []   # list of lists — one per sentence
    clean_sentences = []

    for text in text_list:
        tokens, clean = preprocess_text(text)
        all_tokens.append(tokens)       # .append() keeps tokens separate per sentence
        clean_sentences.append(clean)

    return {
        "tokens":          all_tokens,
        "clean_sentences": clean_sentences
    }


output = full_pipeline(test_sentences)
print(output)
```
