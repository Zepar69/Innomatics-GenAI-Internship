```python
#Unique words in a sentence

sentence=input("Enter your sentence:")
unique=set(sentence.split())
print(f"Unique word count:{len(unique)}")
print(f"Unique words are:{unique}")
```

```python
#Highest salary from emp data

emp={"Ravi":75000,"Anita":28000,"Kiran":72000}
high_emp=max(emp,key=emp.get)
high_sal=emp[high_emp]
print(f"The highest salary is: {high_emp} - {high_sal} ")
```

```python
#Find max and min values

numbers=[45,22,89,10,66]
max=numbers[0]
min=numbers[0]
for num in numbers:
    if num>max:
        max=num
    if num<min:
        min=num
print(f"List:{numbers}")
print(f"Max:{max}")
print(f'Min:{min}')
```

```python
#Count product above price threshold

prices=[450,1200,899,1500,300]
count=0
for price in prices:
    if price>1000:
        count+=1
print(f"Products above 1000:{count}")
```

```python
#Attendance percentage

attendance=["P","P","A","P","P"]

present=attendance.count("P")
total_days=len(attendance)
percentage=(present/total_days)*100

print(f"Attendance Percentage:{percentage}")
```

```python
#Remove duplicate phone numbers

ph_no=[9876543210, 9123456789, 9876543210]

unique_nos=set(ph_no)

print(f"Unique phone numbers: {unique_nos}")
```

```python
#Count character freq

txt="serendipity"
char_count={}
for char in txt:
    if char in char_count:
        char_count[char]+=1
    else:
        char_count[char]=1
print(char_count)
```
